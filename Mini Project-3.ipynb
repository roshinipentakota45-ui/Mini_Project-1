```python
# categorical columns
categorical_cols = [
    "cp", "fbs", "restecg",
    "exang", "slope", "ca", "thal"
]

# Numerical columns
numerical_cols = [
    "age", "trestbps", "chol",
    "thalach", "oldpeak", "sex"
]
```

## ENCODING

```python
encoded_df = pd.get_dummies(df,columns=categorical_cols,drop_first=True)
```

```python
print("Original shape:", df.shape)
print("Encoded shape:", encoded_df.shape)
```

```python
encoded_df.head()
```

```python
bool_cols = encoded_df.select_dtypes(include='bool').columns
encoded_df[bool_cols] = encoded_df[bool_cols].astype(int)

encoded_df.head()
```

## STANDARDIZED

```python
num_cols = ["age", "trestbps", "chol", "thalach", "oldpeak"]

scaler = StandardScaler()
encoded_df[num_cols] = scaler.fit_transform(encoded_df[num_cols])

encoded_df.head()
```

## TRAIN_TEST_SPLIT

```python
x=encoded_df.drop(columns=['target'])
y=encoded_df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```
